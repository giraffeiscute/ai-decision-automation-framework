# random 之資料計算

```python
import numpy as np
import os

# 基礎路徑
base_path = r"C:\Users\Yuan\Desktop\YUAN\OHCA_TBSI_framwork\python-ML-project-Framework-AED-predict-then-optimize\Optimizor\results_random_20250822"

# 不同的 coverage 值
coverages = [5, 10, 20, 40, 60, 80, 100]

# 儲存結果
results = {}

print("=" * 60)
for coverage in coverages:
    # 建立檔案名稱
    file_name = f"random_choose_coverage_{coverage}_set_10.npy"
    file_path = os.path.join(base_path, file_name)
    
    # 檢查檔案是否存在
    if os.path.exists(file_path):
        # 讀取資料
        data = np.load(file_path, allow_pickle=True)
        
        # 計算平均值
        mean_value = np.mean(data)
        
        # 儲存結果
        results[coverage] = mean_value
        
        print(f"Coverage {coverage:3d}: 平均值 = {mean_value:.6f}")
        print(f"  - 資料形狀: {data.shape}")
        print(f"  - 資料範圍: [{np.min(data):.6f}, {np.max(data):.6f}]")
        print("-" * 60)
    else:
        print(f"Coverage {coverage:3d}: 檔案不存在！")
        print(f"  - 路徑: {file_path}")
        print("-" * 60)

print("=" * 60)
print("\n摘要:")
print(f"{'Coverage':<12} {'平均值':<15}")
print("-" * 27)
for coverage, mean_val in results.items():
    print(f"{coverage:<12} {mean_val:<15.6f}")
```

# PIO 之資料計算

```python
import numpy as np
import os

# 基礎路徑
base_path = r"C:\Users\Yuan\Desktop\YUAN\OHCA_TBSI_framwork\python-ML-project-Framework-AED-predict-then-optimize\Optimizor\results_xgb_20250415\shap"

# 不同的 coverage 和 dmin 值
coverages = [5, 10, 20, 40, 60, 80, 100]
dmin_values = [0.6, 0.8, 0.96, 1, 1.2]

# 儲存結果
results = {}

print("=" * 80)
print("開始分析檔案...")
print("=" * 80)

for coverage in coverages:
    results[coverage] = {}
    
    for dmin in dmin_values:
        # 建立檔案名稱 (修正: 使用 coverage 和 dmin 的具體值)
        file_name = f"deployment_decision_coverage_{coverage}_set_10_dist_{dmin}.npy"
        file_path = os.path.join(base_path, file_name)
        
        # 檢查檔案是否存在
        if os.path.exists(file_path):
            try:
                # 讀取資料
                data = np.load(file_path, allow_pickle=True)
                
                # 計算平均值
                mean_value = np.mean(data)
                
                # 儲存結果
                results[coverage][dmin] = {
                    'mean': mean_value,
                    'shape': data.shape,
                    'min': np.min(data),
                    'max': np.max(data),
                    'std': np.std(data)
                }
                
                print(f"✓ Coverage {coverage:3d}, Dmin {dmin:.2f}:")
                print(f"    平均值 = {mean_value:.6f}, 標準差 = {np.std(data):.6f}")
                print(f"    形狀 = {data.shape}, 範圍 = [{np.min(data):.6f}, {np.max(data):.6f}]")
                
            except Exception as e:
                print(f"✗ Coverage {coverage:3d}, Dmin {dmin:.2f}: 讀取錯誤 - {e}")
                results[coverage][dmin] = None
        else:
            print(f"✗ Coverage {coverage:3d}, Dmin {dmin:.2f}: 檔案不存在")
            results[coverage][dmin] = None
    
    print("-" * 80)

print("=" * 80)
print("\n摘要表格:")
print("=" * 80)

# 建立表格標題
header = f"{'N':<12}"
for dmin in dmin_values:
    header += f"{'Dmin=' + str(dmin):<15}"
print(header)
print("-" * (12 + 15 * len(dmin_values)))

# 輸出表格內容
for coverage in coverages:
    row = f"{coverage:<12}"
    for dmin in dmin_values:
        if results[coverage][dmin] is not None:
            row += f"{results[coverage][dmin]['mean']:<15.6f}"
        else:
            row += f"{'N/A':<15}"
    print(row)

print("=" * 80)

# 額外統計: 找出最佳組合
print("\n最佳組合 (最大平均值):")
max_mean = -np.inf
best_coverage = None
best_dmin = None

for coverage in coverages:
    for dmin in dmin_values:
        if results[coverage][dmin] is not None:
            current_mean = results[coverage][dmin]['mean']
            if current_mean > max_mean:
                max_mean = current_mean
                best_coverage = coverage
                best_dmin = dmin

if best_coverage is not None:
    print(f"Coverage = {best_coverage}, Dmin = {best_dmin}, 平均值 = {max_mean:.6f}")
else:
    print("未找到有效數據")
```

```python
import numpy as np
import os
import csv

# 不同的 N 值和 dist 值
N_values = [5, 10, 20, 40, 60, 80, 100]
dist_values = [0.6, 0.8, 0.96, 1, 1.2]

# 基礎路徑
base_path_nn_shap = r"C:\Users\Yuan\Desktop\YUAN\OHCA_TBSI_framwork\python-ML-project-Framework-AED-predict-then-optimize\Optimizor\results_nn_20250415\mlp"
base_path_xgb_shap = r"C:\Users\Yuan\Desktop\YUAN\OHCA_TBSI_framwork\python-ML-project-Framework-AED-predict-then-optimize\Optimizor\results_xgb_20250415\mlp"


# 輸出 CSV
output_csv = "matrix_PO_cov.csv"

# 建立矩陣，第一列是表頭
matrix = [["N值 \\ dist值"] + dist_values]

for N in N_values:
    row = [N]
    for dist in dist_values:
        file_template = f"deployment_decision_coverage_{N}_set_10_dist_{dist}.npy"
        
        try:
            data_nn = np.load(os.path.join(base_path_nn_shap, file_template), allow_pickle=True)
            avg_nn = np.mean(data_nn)
        except:
            avg_nn = None
        
        try:
            data_xgb = np.load(os.path.join(base_path_xgb_shap, file_template), allow_pickle=True)
            avg_xgb = np.mean(data_xgb)
        except:
            avg_xgb = None
        
        if avg_nn is not None and avg_xgb is not None:
            rel_diff = ((avg_xgb - avg_nn) / avg_nn) * 100
            row.append(f"{rel_diff:+.2f}%")
        else:
            row.append("缺少數據")
    
    matrix.append(row)

# 存成 CSV
with open(output_csv, mode="w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerows(matrix)

print(f"\n✅ 已輸出 7×5 的矩陣表到 {output_csv}")
```
